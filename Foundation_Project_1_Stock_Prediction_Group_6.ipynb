```python
#define a function to map POS tags from treebank tag into wordnet tags
wnl = WordNetLemmatizer()
def get_wordnet(pos):
    if pos.startswith('N'):
        return wordnet.NOUN
    elif pos.startswith('V'):
        return wordnet.VERB
    elif pos.startswith('J'):
        return wordnet.ADJ
    elif pos.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN     #default to noun if no match


# Define a function to tag and lemmatize a text string
def lemmatizer(text):
    pos_tags = pos_tag(text)
    text = [wnl.lemmatize(word, pos = get_wordnet(pos)) for word, pos in pos_tags]
    return text

df_english['text_lemmatized'] = df_english['text_tokenized'].apply(lambda x: lemmatizer(x))
```

```python
from datetime import datetime, timedelta

# Define the function to handle flexible date formats
def time_formatting(text):
    # Check if the text contains 'ago'
    if 'ago' not in text:
        # Replace 'nd', 'st', 'rd', 'th' with empty string to handle formats like '2nd', '31st', '4th', etc.
        cleaned_text = text.replace('nd', '').replace('st', '').replace('rd', '').replace('th', '')
        # Parse the datetime string
        dt_obj = datetime.strptime(cleaned_text, '%I:%M %p %b %d')

        # Set the year based on the month
        if dt_obj.month == 1:  # January
            dt_obj = dt_obj.replace(year=2024)
        else:  # February to December
            dt_obj = dt_obj.replace(year=2023)

        return dt_obj
    else:
      parts = text.split()
      hours = 0
      mins = 0
      if 'hours' in text:
            hours = int(parts[0])
            mins = int(parts[2])
      else:
            mins = int(parts[0])

        # Compute current time
      current_time = datetime.now()

        # Subtract hours and minutes from current time
      new_time = current_time - timedelta(hours=hours, minutes=mins)

      return new_time

# Apply the function to create a new column 'new'
train['DateTime'] = train['DateTime'].apply(time_formatting)


```

```python
#Chunking
import spacy
import pandas as pd

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")
```

```python
# Define a custom pattern for NP (Noun Phrase) chunking
def custom_chunking(tags):
    noun_phrases = []
    current_chunk = []
    for token, pos in tags:
        if pos.startswith('N'):
            current_chunk.append(token)
        elif current_chunk:
            noun_phrases.append(current_chunk)
            current_chunk = []
    if current_chunk:
        noun_phrases.append(current_chunk)
    return noun_phrases
```

```python
# Apply chunking to the 'pos_tags' column
train['noun_phrases'] = train['pos_tags'].apply(custom_chunking)
```

```python
#PCA & K-Means Clustering
# Combine sentiment scores and TF-IDF vectors
X = pd.concat([train['compound_score'], tfidf_df], axis=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Step 1: Determine the Number of Components for PCA
explained_variance_threshold = 0.95  # Set your desired threshold

# Initialize PCA
pca = PCA()

# Fit PCA to your TF-IDF matrix
pca.fit(X)

# Calculate explained variance
explained_variance = pca.explained_variance_ratio_

# Calculate cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance)

# Find the optimal number of components
optimal_components = np.argmax(cumulative_explained_variance >= explained_variance_threshold) + 1
```

```python
optimal_components
```

```python
# Step 2: Perform PCA
pca = PCA(n_components=optimal_components)
X_pca = pca.fit_transform(X)
```

```python
# Step 3: Determine the Number of Clusters for K-Means
# Use the Elbow Method to find the optimal number of clusters
inertia = []
max_clusters = 10  # Adjust the maximum number of clusters as needed

for i in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_pca)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(range(1, max_clusters + 1), inertia, marker='o')
plt.title("Elbow Method for Optimal Number of Clusters (K)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
# Choose the optimal number of clusters based on the Elbow Method
optimal_clusters = 3  # Set the optimal number of clusters

# Step 4: Perform K-Means Clustering
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X_pca)
```

```python
# Component 1: Inverted Mean Squared Error (higher is better)
mse = mean_squared_error(y_test, y_pred)
inv_mse = 1 / mse

# Component 2: Hypothetical Profitability
profitability = np.sum((y_pred - y_test) / y_test)

# Component 3: Inverted Risk (Standard Deviation)
risk = np.std(y_pred)
inv_risk = 1 / risk

# Normalizing components (assuming all components are positive)
max_inv_mse = np.max([inv_mse, 1])  # Assuming 1 as a max reference value
max_profitability = np.max([profitability, 1])
max_inv_risk = np.max([inv_risk, 1])

normalized_mse = inv_mse / max_inv_mse
normalized_profitability = profitability / max_profitability
normalized_risk = inv_risk / max_inv_risk

# Weights (sum should be 1)
w_mse = 0.4
w_profitability = 0.4
w_risk = 0.2

# Compound Score
compound_score = (w_mse * normalized_mse +
                  w_profitability * normalized_profitability +
                  w_risk * normalized_risk)

print(f'Compound Score: {compound_score}')
```

```python
import warnings
from sklearn.exceptions import DataConversionWarning

# Suppress specific sklearn warnings
warnings.filterwarnings(action='ignore', category=UserWarning)

import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the data
df = pd.read_csv('/content/Shareholding_Pattern.csv')

# Remove the percentage sign and convert to float
fii_series = df.iloc[1, 1:].str.rstrip('%').astype('float')

# Create a DataFrame for regression with quarters as a numerical value
fii_df = pd.DataFrame({
    'Quarter': np.arange(len(fii_series)),
    'FII': fii_series.values
})

# Set up the regression model
X = fii_df[['Quarter']]  # Independent variable
y = fii_df['FII']  # Dependent variable

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Predict the next quarter
next_quarter = np.array([[X['Quarter'].iloc[-1] + 1]])
predicted_fii = model.predict(next_quarter)

# Print the predicted value
# When predicting, create a DataFrame with the same structure as the fitting data
predict_df = pd.DataFrame(next_quarter, columns=['Quarter'])

# Use the DataFrame to predict
predicted_fii = model.predict(predict_df)

# Print the predicted value
print("Predicted FII for the next quarter:", predicted_fii[0])
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load the data from the CSV file
df = pd.read_csv('/content/ratios.csv')

# Extract the 'Cash Conversion Cycle' data
cash_conversion_cycle = df[df['Unnamed: 0'] == 'Cash Conversion Cycle'].iloc[0, 1:].astype(float)

# Create a DataFrame for regression analysis
years = np.arange(len(cash_conversion_cycle))

# Create the regression model
model = LinearRegression()

# Reshape years for fitting the model
X = years.reshape(-1, 1)
y = cash_conversion_cycle.values

# Fit the model
model.fit(X, y)

# Predict the cash conversion cycle ratio for the next year
next_year = np.array([[len(years)]])
predicted_cycle = model.predict(next_year)

# Print the predicted cycle for the next year
print("Predicted Cash Conversion Cycle for the next year:", predicted_cycle[0])
```
